# Определения

Это файлик с определениями и формулками. На него я буду ссылаться в основном исследовании. Можно сказать это теоретический справочник

## 1. Теоретическая цена фючерса

Теоретическая цена фьючерса рассчитывается по следующим формулам (мы не учитываем дивиденды, потому что до их примерного появления в инфополе еще около полугода): 

$$Упрощенная: \ \ Price_{futures} = Price_{spot} \ \times \ (1 \ + \ \cfrac{r \ \times \ t}{360})$$

$$С \ экспонентой: \ \ Price_{futures} = Price_{spot} \ \times \ e^{\cfrac{r \ \times \ t}{360}}$$

Где 

$Price_{futures} \ - \ теоретическая  \ цена  \ фьючерса$

$Price_{spot} \ - \ цена \  базового  \ актива  \ (акции)$

$r \ - \ безрисковая  \ процентная \  ставка  \ в  \ долях$

$e \ - \ экспонента$

$t \ - \ количество \  дней  \ до  \ экспирации \  фьючерса$

## 2. Ставка безриска

Возьмем за ставку безриска:

$$r_{безриск} \ = \ ключевая \ ставка \ - \ 0.5 \% $$

В моем случае это 20,5 %.

Именно под эту ставку я всегда могу положить деньги с возможностью снятия их в любой день

## 3. Идея арбитража

Есть разные способы проинвестировать в рост компании. Можно купить, собственно, акцию, а можно купить фьючерс на эту акцию. Мы не будем подробно останавливаться на понятии этих инструментов, нас интересует лишь "спекулятивная разница".

Чтобы купить акций на 100 т.р., нужно потратить 100 т.р. Но можно купить фьючерс с номиналом 100 т.р., и тогда заблокируется только гарантийное обеспечение - какая-то сумма меньше 100 т.р. (как правило).

Здесь фьючерс выступает аналогом плеча - заходя в позицию на 100 т.р., у инвестора блокируется 15-40 т.р. (обычно).

Если размышлять с позиции институциональных инвесторов, у них есть, по сути, два инструмента для инвестиций с плечом:

1 - купить фьючерс

2 - купить акции с фондированием через РЕПО

Мы не можем останавливаться на этих понятиях слишком подробно. Отсылаю к лекциям курса "Финансовые рынки" [про РЕПО](https://youtu.be/OrtC1S6-r7M?si=F06GXgrIaG650dmc) и 
[фьючерсы](https://youtu.be/BYK03S2XHmg?si=i3k69u6e82DVUgQj).

Скажем лишь, что в общем случае в разницу цен акции и фьючерса вложена ставка РЕПО = ставка безриска.

Возможность для арбитража возникает, когда цена фьючерса на акцию отличается от справедливой, рассчитанной с учетом стоимости денег и дивидендов. Это позволяет получить арбитражную прибыль при возврате цен к справедливому уровню.

Есть ощущение, что внутри дня цены инструментов могут отклоняться сильнее, чем предполагается в теории, и мне хватит места, чтобы получать арбитржаную прибыль. **Это моя главная гипотеза**

Назовем лонгом "синтетической облигацией" покупка акций и продажа фьючерса - то есть мы по сути покупаем именно разницу в цене инструментов, а шортом "синтетической облигации" - лонг фьючерса и продажу акций - здесь мы продаем эту разницу.

Если реальная цена фьючерса выше теоретический, мы лонгуем корзину:

$$short\ фьючерса + long\ акции\ с\ фондированием\ через\ РЕПО$$

Получая бесплатную прибыль в размере:

$$базис\ - rate_{РЕПО}$$

Теория говорит нам, что мы как физики не сможем получать этот вид арбитражной прибыли, потому что не можем фондироваться через РЕПО, мы можем лишь зайти в акции с плечом, платя брокеру ставку за маржинальное кредитование (ставка у брокера обычно ключ +7-8%, а ставка РЕПО +- ключ). Из-за такой большой ставки базис ее просто не окупит.
**Можно ли заниматься арбитражем в течение дня, так, чтобы у меня не списывались деньги за маржинальное кредитование?**

Если реальная цена фьючерса ниже теоретической, мы шортим корзину:

$$short\ акции \ с \  фондированием \ через \ РЕПО + long\ фьючерса$$

Получая бесплатную прибыль в размере:

$$rate_{РЕПО} -  базис$$

Теория говориит, что я и эту прибыль не смогу получать, так как мне не доступно получение РЕПО ставки. **Могу ли я заработать в этом случае?**

## 4. Допущения

1) Мы покупаем и продаем инструменты с помощью ррыночных заявок. Причем наш объем маленький, поэтому считаем, что заявку исполнят моментально по цене открытия

2) При заключении фьючерсного контракта у нас блокируется ГО, однако это могут быть не деньги, а активы, вложенные, например, под безриск. Поэтому в рассчетах я не учитываю блокировку ГО - у меня нет издержек блокировки денег

3) Ставка безриска для меня = 20,5% годовых

## 5. Комиссии


Комиссия за вход в акции = 0,05% от суммы, а для фьючерса = 0,04% от номинала.
