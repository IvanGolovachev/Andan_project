# Суть исследования

Есть разные способы проинвестировать в рост компании. Можно купить, собственно, акцию, а можно купить фьючерс на эту акцию. Мы не будем подробно останавливаться на понятии этих инструментов, нас интересует лишь "спекулятивная разница".

Чтобы купить акций на 100 т.р., нужно потратить 100 т.р. Но можно купить фьючерс с номиналом 100 т.р., и тогда заблокируется только гарантийное обеспечение - какая-то сумма меньше 100 т.р. (как правило).

Здесь фьючерс выступает аналогом плеча - заходя в позицию на 100 т.р., у инвестора блокируется 15-40 т.р. (обычно).

Если размышлять с позиции институциональных инвесторов, у них есть, по сути, два инструмента для инвестиций с плечом:

1 - купить фьючерс

2 - купить акции с фондированием через РЕПО

Мы не можем останавливаться на этих понятиях слишком подробно. Отсылаю к лекциям курса "Финансовые рынки" [про РЕПО](https://youtu.be/OrtC1S6-r7M?si=F06GXgrIaG650dmc) и [фьючерсы](https://youtu.be/BYK03S2XHmg?si=i3k69u6e82DVUgQj).

Скажем лишь, что в общем случае в разницу цен акции и фьючерса вложена ставка РЕПО.

Возможность для арбитража возникает, когда цена фьючерса на акцию отличается от справедливой, рассчитанной с учетом стоимости денег и дивидендов. Это позволяет получить арбитражную прибыль при возврате цен к справедливому уровню.

Назовем лонгом "синтетической облигацией" покупка акций и продажа фьючерса - то есть мы по сути покупаем именно разницу в цене инструментов, а шортом "синтетической облигации" - лонг фьючерса и продажу акций - здесь мы продаем эту разницу.

Если реальная цена фьючерса выше теоретический, мы лонгуем корзину:

$$short\ фьючерса + long\ акции\ с\ фондированием\ через\ РЕПО$$

Получая бесплатную прибыль в размере:

$$базис\ - rate_{РЕПО}$$

Теория говорит нам, что мы как физики не сможем получать этот вид арбитражной прибыли, потому что не можем фондироваться через РЕПО, мы можем лишь зайти в акции с плечом, платя брокеру ставку за маржинальное кредитование (ставка у брокера обычно ключ +7-8%, а ставка РЕПО +- ключ). Из-за такой большой ставки базис ее просто не окупит.
**Это первое утверждение, которое я хочу проверить.**

А если деньги достаются мне бесплатно? Например, я сниму их с кредитки с полугодовым беспроцентным периодом? Смогу ли я заработать здесь больше, чем ставка безриска? **Это второе утверждение, которое я хочу проверить**


Если реальная цена фьючерса ниже теоретической, мы шортим корзину:

$$short\ акции с фондированием через РЕПО + long\ фьючерса$$

Получая бесплатную прибыль в размере:

$$rate_{РЕПО} -  базис$$

Теория говориит, что я и эту прибыль не смогу получать, так как мне не доступно получение РЕПО ставки. **Это третье утверждение, которое я хочу проверить**

Есть ощущение, что внутри дня цены инструментов могут отклоняться сильнее, чем предполагается в теории, и мне хватит места, чтобы получать арбитржаную прибыль с помощью различных махинаций типа фондирования через бесплатную кредитку. А может и без их помощи - **это тоже вопрос**

**Допущения**

Для начала мы разделим данные на два типа: волатильные и не волатильные. По опыту могу сказать, что профессионалы отключают ботов в волатильные дни (дивиденды, объявление ключевой ставки, пресс-конференции политиков и т.д). Посмотрим, можем ли мы заработать в обычные дни, когда ничего не происходит, и в дни, когда арбитражеров на рынке поменьше - волатильные дни.

Важным моментом будет учет брокерских и ассиметричных биржевых комиссий за мейкерские и тейкерские сделки. Я буду работать через броекра ФИНАМ.

Работать мы будем на российском рынке акций: прогоним модельку и по самым ликвидным, и по не самым.

Так как проект я аппроксимирую на себя, игровые суммы будут небольшие. Это позволит нам не влиять на цену, то есть проводить анализ на старых данных, считая, что наша заявка будет исполнена.
